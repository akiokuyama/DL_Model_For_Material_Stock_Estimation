```python
import csv
import pandas as pd
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Read the based csv file
Base_df = pd.read_csv("/Users/akihi/Library/CloudStorage/OneDrive-PrincetonUniversity/Projects/MS Estimation/Results/MS_Estimation/Merged.csv")
Base_df
```

```python
# Create subgroup dataset
Non_R = Base_df[Base_df['BldType_True'] == 0]
SFH_df = Base_df[Base_df['BldType_True'] == 1]
MFH_df = Base_df[Base_df['BldType_True'] == 2]

# Check if I could sucessfully create the subsets
unique_Non_R = Non_R['BldType_True'].unique()
unique_SFH_df = SFH_df['BldType_True'].unique()
unique_MFH_df = MFH_df['BldType_True'].unique()
print("Unique values in Non_R:", unique_Non_R)
print("Unique values in SFH_df:", unique_SFH_df)
print("Unique values in MFH_df:", unique_MFH_df)
```

```python
# Non_Residential
y_true_Non_R = Non_R['FA_True_scaled']
y_pred_Non_R = Non_R['FA_Pred_scaled']

y_true_Non_R_unscaled = Non_R['FA_True']
y_pred_Non_R_unscaled = Non_R['FA_Pred']

# Calculating MSE
mse = mean_squared_error(y_true_Non_R, y_pred_Non_R)

# Calculating RMSE
rmse = np.sqrt(mse)

# Calculating MAE
mae = mean_absolute_error(y_true_Non_R, y_pred_Non_R)

# Calculating R2 Score
r2 = r2_score(y_true_Non_R, y_pred_Non_R)

# Calculating error between total floor area
Dif = np.abs(np.sum(y_true_Non_R_unscaled) - np.sum(y_pred_Non_R_unscaled))
Error = (Dif/np.abs(np.sum(y_true_Non_R_unscaled))) * 100

# Printing the results
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Error (MAE):", mae)
print("R-squared (R2):", r2)
print("Error between total floor area:", Error)
```
