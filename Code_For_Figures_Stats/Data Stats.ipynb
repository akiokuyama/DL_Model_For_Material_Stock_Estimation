```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Grid_MR = pd.read_csv("/scratch/network/ao3526/MR_2011_2021withBldType.csv")

Grid_MR.rename(columns={'ID': 'id'}, inplace=True)
Grid_MR.rename(columns={'FA_chng': 'FA_Change'}, inplace=True)
Grid_MR.rename(columns={'latitud': 'latitude'}, inplace=True)
Grid_MR.rename(columns={'longttd': 'longitude'}, inplace=True)

Grid_MR['id'] = Grid_MR['id'].astype(int)
Grid_MR
```

```python
Grid_MR.rename(columns={'FA22021': 'FA2021'}, inplace=True)


fa_min = Grid_MR['FA2021'].min()
fa_max = Grid_MR['FA2021'].max()

print("Minimum FA_Change:", fa_min)
print("Maximum FA_Change:", fa_max)
```

```python
# Visualize the distribution of the "FA_Change" column
plt.figure(figsize=(10, 6))
plt.hist(Grid_MR['FA2021'], bins=20, color='skyblue', edgecolor = 'black')
plt.title("Distribution of FA2021")
plt.xlabel("FA2021")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=Grid_MR['FA2021'])
plt.title("Box Plot of FA2021")
plt.xlabel("FA2021")
plt.show()

# Descriptive statistics
print(Grid_MR['FA2021'].describe())
```

```python
fa_change_min = Grid_MR['FA_Change'].min()
fa_change_max = Grid_MR['FA_Change'].max()

print("Minimum FA_Change:", fa_change_min)
print("Maximum FA_Change:", fa_change_max)
```

```python
# Visualize the distribution of the "FA_Change" column
plt.figure(figsize=(10, 6))
plt.hist(Grid_MR['FA_Change'], bins=20, color='skyblue', edgecolor = 'black')
plt.title("Distribution of FA_Change")
plt.xlabel("FA_Change")
plt.ylabel("Frequency")
plt.show()
```

```python
# KDE plot
plt.figure(figsize=(10, 6))
sns.kdeplot(Grid_MR['FA_Change'], color="blue", shade=True)
plt.title("KDE of FA_Change")
plt.xlabel("FA_Change")
plt.ylabel("Density")
plt.show()

# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=Grid_MR['FA_Change'])
plt.title("Box Plot of FA_Change")
plt.xlabel("FA_Change")
plt.show()

# Descriptive statistics
print(Grid_MR['FA_Change'].describe())
```
